```python
import dask.dataframe as dd
from dask.diagnostics import ProgressBar 
from matplotlib import pyplot as plt 
df = dd.read_csv('./*2017.csv', dtype={'House Number': 'object', 'Issuer Squad': 'object', 'Violation Post Code': 'object', 'Time First Observed': 'object'}) 
df
```

```python
# Counting missing values in DF
missing_values = df.isnull().sum()
missing_values
```

```python
# Calculating the percent of missing values in DF
missing_count = ((missing_values / df.index.size) * 100)
missing_count
```

```python
# Computing the DAG
with ProgressBar():
  missing_count_pct = missing_count.compute()
missing_count_pct
```
